```python
# 26.8206, 30.8025 lat and long of egypt
from geopy.distance import great_circle
```

```python
def ext_distance(x):
    try:
        return great_circle((x['Lat'],x['Long']),('26.8206','30.8025')).kilometers
    except:
        return np.nan

data['Dist_From_Egy']=data.apply(ext_distance, axis=1)
data
```

```python
df4=pd.read_csv('Restaurant_Reviews.csv', sep='\t')
```

```python
df4
```

#### 1) Using Restaurant_Reviews.csv dataset extract Bag Of Words CountVectorizer Features.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer(stop_words='english')

vector=vectorizer.fit_transform(df4['Review'])
vector.toarray()
```

```python
data_res=pd.DataFrame(vector.toarray(),columns=vectorizer.get_feature_names())
data_res
```

#### 2) Using Restaurant_Reviews.csv dataset extract TfidfVectorizer Features.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words='english')
vector=vectorizer.fit_transform(df4['Review'])
vector.toarray()
```

```python
data_resp2=pd.DataFrame(vector.toarray(),columns=vectorizer.get_feature_names())
data_resp2
```
