## Answer Questions using SQL Language

```python
import sqlite3
connection=sqlite3.connect('Books.db')
rows=connection.execute('Select * from Books where Book_Rate >= ? and Book_Name like ? ', [3,'%Mr%']).fetchall()
```

```python
#Query1
rows
print('there are {} rows that meet the condition of the query'.format(len(rows)))
for row in rows:
    print(row)
```

```python
#Query2
rows=connection.execute('Select * from Books where Book_Rate >= ? and Book_Price > ? ', [3,20]).fetchall()
rows
```

```python
#Query3
rows=connection.execute('Select * from Books where Book_Rate IN (?,?,?,?) ', [1,2,4,5]).fetchall()
rows
len(rows)
```

```python
#Query4
rows=connection.execute('Select * from Books where (Book_Price Between ? and ?) AND (Book_Rate IN (?,?,?)) ', [10,40,3,4,1]).fetchall()
rows
len(rows)
```

```python
#Query5
rows=connection.execute('SELECT * FROM Books order by Book_Price Desc limit ?',[5]).fetchall()
```

```python
#Query6
rows=connection.execute('SELECT * FROM Books order by Book_Rate Desc, Book_Price ASC Limit ? offset ? ', [10,20]).fetchall()
rows
```

```python
# Query 7
rows=connection.execute('Insert into Category (Category_Name) values(?)', ['Francine Rivers'])
connection.commit()
#Francine Rivers records data
Francine_Rivers_idx=list(range(953,958))
Francine_Rivers_names=['Redeeming Love','A Voice In The Wind','The Masterpiece','Her Mother\'s Hope','Her Daughter\'s Dream']
Francine_Rivers_Price=[100,80,90,120,110]
Francine_Rivers_Rate=[5,4,5,3,2]
Francine_Rivers_Categ_ID=[49]*5
for idx,name,price,rating,category_id in zip(Francine_Rivers_idx,Francine_Rivers_names,Francine_Rivers_Price,Francine_Rivers_Rate,Francine_Rivers_Categ_ID):
    connection.execute('INSERT INTO Books (Book_ID,Book_Name,Book_Price,Book_Rate,Category_ID)VALUES(?,?,?,?,?)',[idx,name,price,rating,category_id])
    connection.commit()
```

```python
#Query 8
rows=connection.execute('Update Books set Book_Rate = ? where Book_Price < ?', [3,20]).fetchall()
connection.commit()
```

```python
#query 9
rows=connection.execute('Delete from Books where Book_Rate <= ? and Book_Price > ?', [2,50]).fetchall()
connection.commit()
```

```python
#Query14
rows=connection.execute('SELECT b.Book_Name,c.Category_name, b.Book_Price,b.Book_Rate from Books b inner join Category c On b.Category_ID = c.Category_id').fetchall()
rows
```

```python
#Query 15
rows=connection.execute('SELECT count(Book_ID), Book_rate from books where Book_Price between ? and ? group by Book_Rate', [20,30]).fetchall()
rows
for count,rate in rows:
    print('The number of books with rating {} star is {} between price 20 and 30 '.format(rate,count))
```

```python
#Query 16
rows=connection.execute('SELECT count(Book_ID) , c.Category_name from Books b inner join Category c On b.Category_ID = c.Category_id group by c.Category_name having count(Book_ID)>10').fetchall()
rows
for counts, category in rows:
    print('The number of books in the {} category is {}'.format(category,counts))
```

```python
#query 17
rows=connection.execute('SELECT * from books where Category_ID= (select Category_id from Category where Category_name="Music" ) ').fetchall()
rows
for i in rows:
    print(i[1])
```
