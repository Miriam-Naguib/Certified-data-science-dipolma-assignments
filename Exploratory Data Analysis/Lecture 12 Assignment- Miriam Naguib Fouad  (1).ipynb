```python
import pandas as pd
df=pd.read_csv('bikes.csv')
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1- Profit feature

```python
df['Profit'] = (df['registered']*5 + df['casual']*20)*0.86-(1500/(365*24))
df
```

# 3- Profit for each day.

```python
df['datetime']=pd.to_datetime(df['datetime'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
df
```

```python
df['hour']=df['datetime'].dt.hour
df['day']=df['datetime'].dt.day
df['month']=df['datetime'].dt.month
df['year']=df['datetime'].dt.year
df['week']=df['datetime'].dt.week
df['dayname']=df['datetime'].dt.day_name()
df
```

# 10. Schools (9-12) & (2-6) profit.

```python
def School_Period(x):
    if x in range(9,13):
        return "School_Semester_1 (9-12)"
    elif x in range(2,7):
        return "School_Semester_2 (2-6)"
    else:
        return "School_Vacation"
df['School']=df['month'].apply(School_Period)
```

```python
df
```

```python
groupby_school=df.groupby('School').sum()
groupby_school.drop('School_Vacation',axis=0)
```

```python
groupby_school.drop('School_Vacation',axis=0)[['registered','rented_bikes_count','casual','Profit']]
```

## 7. Rentals during rush hours (7-9 am & 3-5 pm).

```python
def period_ext(x):
    if x in range(7,10):
        return 'Morning Rush Hour, 7-9 am'
    elif x in range(15,18):
        return "Evening Rush Hour, 3-5 pm"
    else:
        return "Normal Hour"
    
df['Period']=df['hour'].apply(period_ext)
df
```

```python
groupby_period=df.groupby('Period').sum()
groupby_period
```

```python
groupby_period.sort_values(by=['rented_bikes_count'], ascending=False)['rented_bikes_count']
groupby_period.drop('Normal Hour', axis=0)
```
